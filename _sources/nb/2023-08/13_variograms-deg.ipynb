```python
def default_props_anesth(
        num_k: int = 7,
        perc: int = 15,
        **kwargs, ):
    mice = Mice(512)
    mice.set_kws(
        desc_ca2='preproc',
        desc_bold='rabies-hp',
    )
    props_base = {
        'mice': mice,
        'num_k': num_k,
        'task': 'rest',
        'perc': f'p{perc}-sample',
        'metric': 'pearson',
        'sv_props': {
            'match_using': 'gam',
            'match_metric': 'euclidean'},
        'graph_type': 'real',
        'kw_kmeans': {'n_init': 100},
        'verbose': False,
    }
    kwargs = setup_kwargs(props_base, kwargs)
    props_ca = {'mode': 'ca2', **kwargs}
    return mice, props_ca


def default_props_awake(
        num_k: int = 7,
        perc: int = 15,
        **kwargs, ):
    mice = Mice(Config(
        nn=512,
        sub_ids=[2, 6, 7, 9, 10],
        ses_ids=[4],
        run_ids=range(1, 10 + 1),
    ))
    mice.set_kws(
        desc_ca2='awake',
        desc_bold='rabies-hp',
    )
    props_base = {
        'mice': mice,
        'num_k': num_k,
        'task': 'rest',
        'perc': f'p{perc}-sample',
        'metric': 'pearson',
        'sv_props': {
            'match_using': 'gam',
            'match_metric': 'euclidean'},
        'graph_type': 'real',
        'kw_kmeans': {'n_init': 100},
        'verbose': False,
    }
    kwargs = setup_kwargs(props_base, kwargs)
    props_ca = {'mode': 'ca2', **kwargs}
    return mice, props_ca


def _get_avg_deg_ca(mice):
    deg = {}
    for b in bands:
        mice.set_kws(band_ca2=b)
        d = collect_degrees_lite(
            mice=mice,
            mode='ca2',
            rho=0.15,
            filter_isolated=True,
        )
        x = np.stack(list(d.values()))
        deg[b] = bn.nanmean(x, 0)
    deg = {
        f"{bands[b]}\n{b}": d
        for b, d in deg.items()
    }
    return deg
```

```python
deg = {
    BOLD: deg_bold,
    CAS: deg_anesth[f"{CAS}\n(0.01, 0.5)"],
    CAF: deg_anesth[f"{CAF}\n(0.5, 5.0)"],
}
```

```python
ids_present_left = mice.parcel.df.loc[
    (mice.parcel.df['hemis'] == 1) &
    (mice.parcel.df['id'].isin(com)),
    'id'
].values
len(ids_present_left)
```

```python
distances_left = {
    (i, j): np.linalg.norm(com[i] - com[j]) for i, j
    in itertools.combinations(ids_present_left, 2)
}
len(distances_left)
```

```python
delta = 50

num = 100
span = np.linspace(
    min(distances_left.values()),
    max(distances_left.values()),
    num=num,
)
gamma = {
    k: np.zeros(num)
    for k in deg
}
for i0, h in enumerate(span):
    neighbors = dict(filter(
        lambda t: h - delta < t[1] < h + delta,
        distances_left.items(),
    ))
    for mode, d in deg.items():
        var = np_nans(len(neighbors))
        for ctr, (i, j) in enumerate(neighbors):
            xi = d[bs.id2n[i]]
            xj = d[bs.id2n[j]]
            var[ctr] = (xi - xj) ** 2
        var = np.nanmean(var)
        gamma[mode][i0] = var
```

```python
pal = prepare_palette()

fig, axes = create_figure(1, 3, (8, 1.6), layout='constrained')
for i, mode in enumerate(gamma):
    axes[i].plot(
        span,
        gamma[mode],
        color=pal[mode],
        label=mode,
        lw=3,
    )
    axes[i].legend()

fig.savefig(
    pjoin(fig_dir, f"variogram_degree.pdf"),
    # bbox_extra_artists=[s],
    bbox_inches='tight',
    # dpi=200,
)
plt.show()
```
