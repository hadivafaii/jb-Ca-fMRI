## Prepare (128)

```python
mice = Mice(128)

mako = sns.color_palette('mako', as_cmap=True)
rocket = sns.color_palette('rocket', as_cmap=True)
pal = {'ca2': mako(0.6), 'bold-lite': rocket(0.6)}
```

```python
task = 'rest'
b = (0.01, 0.5)
mice.setup_func_data(task, band_ca=b)

num_k, p = 7, 20
perc = f'p{p}-sample'
props_base = {
    'kw_kmeans': {'n_init': 100},
    'perc': perc,
    'num_k': num_k,
    'mice': mice,
    'task': task,
    'metric': 'pearson',
    'match_using': 'gam',
    'match_metric': 'euclidean',
    'graph_type': 'real',
    'verbose': False,
}
props_ca = {'mode': 'ca2', **props_base}
props_bo = {'mode': 'bold-lite', **props_base}

bt_ca = BootStrap(**props_ca)
bt_bo = BootStrap(**props_bo)
```

### Fix and resave run deg

#### Bold

```python
for kk in range(3, 25):
    props_bo['num_k'] = kk
    self = Group(**props_bo)
    if not self.fit_meta:
        continue
    print('-' * 70)
    print(self.fit_meta)

    self.load_svinets()
    degree = self.collect_degrees()
    meta_key = f"{self.mode.split('-')[0]}_{self.task}"

    deg_runs = np_nans((
        self.mice.cfg.n_subjects,
        self.mice.cfg.n_sessions,
        max(self.n_runs.values()),
        len(self.n2l),
    ))
    for key, (sub, ses, _) in self.mice.looper_agg.items():
        if sub == -1 or ses == -1:
            continue
        for i, run_k in enumerate(
                self.mice.combine_meta[meta_key][key]):
            if run_k not in self.svinets:
                print(f"Warning! {run_k} was not fit.  Moving on . . .")
                continue
            ids = (sub - 1, ses - 1, i)
            deg_runs[ids] = degree.pop(run_k)

    _ = save_obj(
        obj=deg_runs,
        file_name='run_deg',
        save_dir=self.path,
        verbose=True,
        mode='npy',
    )

    sns.histplot(deg_runs.ravel(), bins=np.linspace(0, 100, 101))
    plt.title(f"K = {kk}")
    plt.show()

    print('-' * 70)
    print('\n\n')
```
