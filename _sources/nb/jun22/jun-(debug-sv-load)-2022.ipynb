```python
%%time

self.perms, _ = self.align_group()
data = self._apply_perms()
```

```python
n_seeds = {}
for key, sv in self.svinets.items():
    n_seeds[key] = len(sv.gam_all)

meta_key = f"{self.mode.split('-')[0]}_{self.task}"
n_runs = {}
for key, (sub, ses, _) in self.mice.looper_agg.items():
    if sub == -1 or ses == -1:
        continue
    n_runs[key] = len(self.mice.combine_meta[meta_key][key])
```

```python
sns.histplot(n_seeds.values());
```

```python
pi_runs = np_nans((
    self.mice.cfg.n_subjects,
    self.mice.cfg.n_sessions,
    max(n_runs.values()),
    max(n_seeds.values()),
    self.num_k,
    len(self.n2l),
))
lam_runs = np_nans((*pi_runs.shape[:-1], 2))
```

```python
address = {}
for key, (sub, ses, _) in self.mice.looper_agg.items():
    if sub == -1 or ses == -1:
        continue

    for i, run_k in enumerate(
            self.mice.combine_meta[meta_key][key]):
        if run_k not in self.svinets:
            if self.verbose:
                print(f"Warning! {run_k} was not fit.  Moving on . . .")
            continue
        ids = range(n_seeds[run_k])
        address[run_k] = (sub - 1, ses - 1, i)
        pi_runs[sub - 1, ses - 1, i][ids] = data[run_k]['pi']
        lam_runs[sub - 1, ses - 1, i][ids] = data[run_k]['lam']
```
